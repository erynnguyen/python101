## 3. Working with Missing Values

**▸ Sources of Missing Values**

Before we dive into code, it’s important to understand the sources of missing data. Here’s some typical reasons why data is missing:
- User forgot to fill in a field.
- Data was lost while transferring manually from a legacy database.
- There was a programming error, (when join, merge tables)
- Users chose not to fill out a field tied to their beliefs about how the results would be used or interpreted.

As you can see, some of these sources are just simple random mistakes. Other times, there can be a deeper reason why data is missing.

**▸ Methods to work with Missing Values**

* Delete the case with missing observations. This is OK if this only causes the loss of a relatively small
number of cases. This is the simplest solution.
* Fill-in the missing value with mean, mode, median or other constant value.
* Use the rest of the data to predict the missing values by regression, KNN, ...
* Missing observation correlation. Consider just
(xi, yi) pairs with some observations missing. The
means and SDs of x and y can be used in the estimate even when a member of a pair is missing.

In Pandas, there are several useful methods for detecting, removing, and replacing null values:

* `isna()` or `isnull()`: Generate a boolean mask indicating missing values
* `notna()` or `notnull()`: Opposite of isnull()
* `dropna()`: Return a filtered version of the data
* `fillna()`: Return a copy of the data with missing values filled or imputed

**▸ Detecting null value**: Pandas data structures have two useful methods for detecting null data: `isnull()` and `notnull()`. Either one will return a Boolean mask over the data. For example:

## 6. Detect outliers with Inter-quantile Range


````python
# Compute a descriptive statistics review
df.describe()

df['Fare'].describe()
````

One of the measure that describes how spread out the data is is the **Quantile**, which divides data into 4 equal quantiles. 

<img src='https://cdn.scribbr.com/wp-content/uploads/2020/09/iqr_quartiles.png'>

**Interquartiles Range** độ chênh lệch giữa 2 quantile

<img src='https://miro.medium.com/max/1400/1*2c21SkzJMf3frPXPAR_gZA.png' width=700>

To identify the outliers (phần tử ngoại lai), we follow 3 steps:

- ***Step 1***: Identify the Inter-quartiles Range 

    IQR = Q3 - Q1


- ***Step 2***: Calculate the lower whisker and upper whisker

    upper-whisker = Q3 + 1.5*IQR

    lower-whisker = Q1 - 1.5*IQR


- ***Step 3***: Identify the outliers. Data that is below the lower whisker or above the upper whisker will be considered to be outliers.

````python
# Visualize with boxplot
import matplotlib.pyplot as plt
df['Fare'].plot(kind='box', figsize=(10, 10))
plt.show()
````

````python
# 1. Calculate the quantiles 
q1 = df['Fare'].quantile(0.25) # Tính Q1
q3 = df['Fare'].quantile(0.75) # Tính Q3 

# 2. Calculate the interquantile range
iqr = q3 - q1
iqr

# # 3. Calculate the whisker: vùng bình thường của dataset nằm trong khoảng upper và lower
upper = q3 + 1.5*iqr
lower = q1 - 1.5*iqr
````

````python
upper
````

````python
lower
````

````python
# Khoảng bình thường
df[(lower < df['Fare']) & (df['Fare'] < upper)]
````

````python
# Vùng ngoại lai --- > upper, < lower
df[(lower > df['Fare']) | (df['Fare'] > upper)]
````

## 3. *Example* - Titanic: Recover Gender column.

While working on the Titanic dataset, you accidentally dropped the column "Sex". From other columns, try to recover the new Sex column.

````python
new = titanic.drop(columns='Sex')
new.head(10)
````

````python
# Recovery the Sex column by checking name's suffix
# Suffix is located between "," and "."

# 1. Try getting suffix on one example of name
name = 'Palsson, Master. Gosta Leonard'
name.split(', ')[1]
name.split(', ')[1].split('.')[0]

# 2. Get suffix of all the row --> Use apply
def get_suffix(name):
    return name.split(', ')[1].split('.')[0]

new['Suffix'] = new['Name'].apply(get_suffix)

# 3. Check how many unique suffixes there are 
new['Suffix'].unique()

# 4. From the Suffix column, compute the Sex column
def check_sex(suffix):
    if suffix in ['Mrs', 'Miss', 'Mme', 'Ms', 'Lady', 'the Countess']:
        return 'female'
    else:
        return 'male'

new['Sex'] = new['Suffix'].apply(check_sex)

# 5. Check the accuracy of our new columns (vs the original column)
(new['Sex'] == titanic['Sex']).mean()
````

````python
new[new['Sex'] != titanic['Sex']]
````

````python
titanic[new['Sex'] != titanic['Sex']]
````

````python
#@title Solution
# ---------- 1. Try getting suffix on one example of name
name = "Nasser, Mrs. Nicholas (Adele Achem)"
name.split(".")[0].split(', ')[1]

# ---------- 2. Get suffix of all the row --> Use apply
def get_suffix(name):
    return name.split(".")[0].split(', ')[1]

new['suffix'] = new['Name'].apply(get_suffix)

# ---------- 3. Check how many unique suffixes there are 
new['suffix'].unique()

# ---------- 4. From the Suffix column, compute the Sex column
female_suffix = ['Mrs', 'Miss', 'Ms']

def check_sex(suffix):
    if suffix in female_suffix:
        return "female"
    else:
        return "male"

new['sex'] = new['suffix'].apply(check_sex)

# ---------- 5. Check the accuracy of our new columns (vs the original column)
(titanic['Sex'] == new['sex']).mean()
````

3. Lowercase and Wrap into a Function

````python
letters.lower()
````

````python
# Wrap them into a function and apply to the dataset 

def clean_text(x):
    # Remove html tag
    raw = BeautifulSoup(x, "html.parser")
    raw = raw.get_text()

    # Remove non-alphabetic characters + lowercase
    clean = re.sub("[^a-zA-Z ]", " ", raw).lower()
    return clean

imdb['clean'] = imdb['review'].apply(clean_text)
````

✳︎ *Application* ⎯ Calculate Word Frequency

````python
# ⎯ Step 1: Innitiate an empty dictionary to save the result. 
# Data to save into this dictionary will have the structure of 
# 'word' as key and its frequency as value.
result_dict = {}

imdb['clean'][0].split()
````

````python
# ⎯ Step 2: Design a function that:
# - Receive a sentence as its input
# - Split the sentence to a list of words
# - Iterate over the list and check
# ⎯ if word is not in the result dictionary yet → add word to the dictionary with frequency = 1
# ⎯ if word is already in the result dictionary → increase its frequency +1  

def count_word(review):
    words = review.split()
    for word in words:
        if word not in result_dict:
            result_dict[word] = 1
        elif word in result_dict:
            result_dict[word] += 1
````

````python
# ⎯ Step 3: Apply the function to the dataset.
# YOUR CODE HERE
imdb['clean'].apply(count_word)
````

## 4. Stemming / Lemmatization

For grammatical reasons, documents are going to use different forms of a word, such as organize, organizes, and organizing. Additionally, there are families of derivationally related words with similar meanings, such as *democracy*, *democratic*, and *democratization*. In many situations, it seems as if it would be useful for a search for one of these words to return documents that contain another word in the set.

The goal of both stemming and lemmatization is to reduce inflectional forms and sometimes derivationally related forms of a word to a common base form. For instance:

- am, are, is $\Rightarrow$ be

- car, cars, car's, cars' $\Rightarrow$ car

The result of this mapping of text will be something like:

- the boy's cars are different colors $\Rightarrow$ the boy car be differ color

Stemming and Lemmatization are two different methods to map text to its root.

- **Stemming**: refers to a crude heuristic process that chops off the ends of words.
- **Lemmatization**:  refers to doing things properly with the use of a vocabulary and morphological analysis of words.
