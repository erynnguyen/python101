```python
from urllib.request import urlopen 
import re

def read_url(url): 
    return re.sub('\\s+', ' ', urlopen(url).read().decode())

text = read_url('https://raw.githubusercontent.com/dhminh1024/practice_datasets/master/huck_finn.txt')
```

```python
def upper(a):
  a = a + '333'
  return a

result = ''.join(map(upper, lista))
print(result)
```
