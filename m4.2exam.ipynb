```python
import pandas as pd

confirmed  = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
deaths     = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
recoveries = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
```

```python
latest_confirmed = confirmed[['Country/Region', '2/9/23']]
latest_confirmed = latest_confirmed.rename(columns={'Country/Region':'Country','2/9/23':'confirmed'})
latest_confirmed
```

```python
latest_confirmed['Country'].nunique()
```

```python
latest_deaths = deaths[['Country/Region', '2/9/23']]
latest_deaths = latest_deaths.rename(columns={'Country/Region':'Country','2/9/23':'deaths'})
latest_deaths
```

```python
latest_recoveries = recoveries[['Country/Region', '2/9/23']]
latest_recoveries = latest_recoveries.rename(columns={'Country/Region':'Country','2/9/23':'recoveries'})
latest_recoveries
```

```python
pd.merge(latest_confirmed, latest_deaths, how='left', on='Country')
```
